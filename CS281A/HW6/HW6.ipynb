```python
from __future__ import division, print_function
import numpy as np
from matplotlib import pyplot as plt
%pylab inline
```

```python
train = np.loadtxt('hmm-gauss.dat')
```

```python
plt.plot(train[:,0],train[:,1],'bo')
```

# Initialize Variables

```python
n_data = train.shape[0]
n_dim = train.shape[1]
n_states = 4
data_mean = np.mean(train,axis=0)
data_std = np.linalg.norm(np.std(train,axis=0))
rng = np.random.RandomState(1)
#mu = rng.randn(n_states,n_dim)
mu = 10.*rng.normal(loc=data_mean,scale=data_std,size=(n_states,n_dim))
#sigma2 = rng.randn(n_states)**2
sigma2 = 10.*np.ones(n_states)*(data_std**2)/n_states
pi = np.array([.25,.25,.25,.25])
#pi = np.squeeze(rng.dirichlet(np.ones(n_states)))
a = rng.dirichlet(np.ones(n_states),size=n_states)
pq = np.zeros(shape=(n_data,n_states))
```

```python
sigma2
```

# Updates

```python
def piUp(pq):
    return pq[0]
def aUp(pq):
    aTemp = np.zeros(shape=(n_states,n_states))
    for ii in xrange(n_states):
        for jj in xrange(n_states):
            aTemp[ii,jj] = np.sum([pq[tt,ii]*pq[tt+1,jj] for tt in xrange(n_data-1)])/np.sum([np.sum([pq[tt,ii]*pq[tt+1,nn] for tt in xrange(n_data-1)]) for nn in xrange(n_states)])
    return aTemp
def muUp(pq,data):
    muTemp = np.zeros(shape=(n_states,n_dim))
    for ii in xrange(n_states):
        muTemp[ii] = pq[:,ii].dot(data)/np.sum(pq[:,ii],axis=0)
    return muTemp
def sigma2Up(pq,data,mu):
    sigma2Temp = np.zeros(shape=(n_states))
    for ii in xrange(n_states):
        sigma2Temp[ii] = pq[:,ii].dot(np.einsum('ij,ij->i',(train-mu[ii]),(train-mu[ii])))/np.sum(pq[:,ii])
    return sigma2Temp/2.

def logAlphaUp(pi,a,data,mu,sigma):
    temp = np.zeros(n_states)
    logAlphat = np.zeros(shape=(n_data,n_states))
    logAlphat[0] = np.log(pi)+logpy(data[0],mu,sigma)
    for tt in xrange(1,n_data):
        for ii in xrange(n_states):
            for jj in xrange(n_states):
                temp[jj] = logAlphat[tt-1,jj]+np.log(a[jj,ii])+logpy(data[tt],mu,sigma)[ii]
            logAlphat[tt,ii] = sumExp(temp)
    return logAlphat
def logBetaUp(pi,a,data,mu,sigma):
    temp = np.zeros(n_states)
    logBetat = np.zeros(shape=(n_data,n_states))
    for tt in xrange(n_data-2,-1,-1):
        for ii in xrange(n_states):
            for jj in xrange(n_states):
                temp[jj] = logBetat[tt+1,jj]+np.log(a[ii,jj])+logpy(data[tt+1],mu,sigma)[jj]
            logBetat[tt,ii] = sumExp(temp)
    return logBetat
def gammaUp(logAlphat,logBetat):
    gammaTemp = np.zeros(shape=(n_data,n_states))
    for ii in xrange(n_data):
        gammaTemp[ii] = np.exp(logAlphat[ii]+logBetat[ii]-sumExp(logAlphat[ii]+logBetat[ii]))
    return gammaTemp
def logpy(data,mu,sigma2):
    tempPy = np.zeros(n_states)
    for ii in xrange(n_states):
        tempPy[ii] = -np.log(2*np.pi*sigma2[ii])+(-.5*np.linalg.norm(data-mu[ii])**2/sigma2[ii])
    return tempPy
def sumExp(things):
    return np.amax(things)+np.log(np.sum(np.exp(things-np.amax(things))))
```

# EM

```python
steps = 10
pinit = pq.copy()
for ii in xrange(steps):
    logAlpha = logAlphaUp(pi,a,train,mu,sigma2)
    logBeta = logBetaUp(pi,a,train,mu,sigma2)
    gamma = gammaUp(logAlpha,logBeta)
    for jj in xrange(n_data):
        pq[jj] = gamma[jj]/np.sum(gamma[jj])
    pi = piUp(pq)
    a = aUp(pq)
    mu = muUp(pq,train)
    sigma2 = sigma2Up(pq,train,mu)
print('Diff: '+str(np.mean(np.abs(pinit-pq))))
```
